```python
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

from sklearn.model_selection import train_test_split
from sklearn import metrics
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
expr = pd.read_csv("data/expresion_matrix.csv")
expr.head()
```

```python
sen = pd.read_csv("data/sensitivity_matrix_Activity_Area.csv", index_col = 0)
sen.head()
```

```python
X = np.array(expr)
```

```python
model = KMeans(n_clusters=3, random_state=0, n_init="auto")
# model = AgglomerativeClustering(n_clusters=3)
kmeans = model.fit(X)
labels = kmeans.labels_
```

```python
k_predefined = [2, 3, 4, 5]
sil_score = []

for k in k_predefined:
    model = KMeans(n_clusters=k, random_state=0, n_init="auto")
    kmeans = model.fit(X)
    labels = kmeans.labels_
    sc = silhouette_score(X, labels).item()
    sil_score.append(sc)
sil_score
```
